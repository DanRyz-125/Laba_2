```python
import urllib.request
import urllib.error
import time
import datetime
import os
def download(index):
    url = 'https://www.star.nesdis.noaa.gov/smcd/emb/vci/VH/get_TS_admin.php?country=UKR&provinceID={}&year1=1981&year2=2020&type=Mean'.format(index)
    vhi_url = urllib.request.urlopen(url)
    date = datetime.datetime.now().strftime("%d_%m_%Y_%H_%M_%S")
    out = open('vhi_id_%02d.csv' %index, 'wb')
    out.write(vhi_url.read())
    out.close()  
    os.rename('vhi_id_%02d.csv' %index, date+'_vhi_id_%02d.csv' %index)
    
for index in range(1, 28):
    download(index)
```

```python
import csv

def clean_this(index):

    file_path = 'C:\\Users\\Home\\LabPython\\downloaded\\15_11_2022_14_vhi_id_%02d.csv'%index
    cleaned_path = 'C:\\Users\\Home\\LabPython\\cleaned\\15_11_2022_14_vhi_id_%02d.csv'%index
    
    with open(file_path, 'r') as old_file:
        reader = csv.reader(old_file, delimiter=',')
        
        with open(cleaned_path, 'w') as new_file:
            f = csv.writer(new_file, delimiter=',')
        
            for line in reader:
                contains_bad=False
                for l in line:
                    try:
                        contains_bad=contains_bad or int(float(l)) == -1
                    except:
                        pass
                if contains_bad:
                    continue
                f.writerow(line)

                
for ident in range (1, 28):
    clean_this(index)
```
