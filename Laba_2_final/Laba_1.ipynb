```python
import urllib.request
import urllib.error
import time
import datetime
import os
def download(index):
    url = 'https://www.star.nesdis.noaa.gov/smcd/emb/vci/VH/get_TS_admin.php?country=UKR&provinceID={}&year1=1981&year2=2020&type=Mean'.format(index)
    vhi_url = urllib.request.urlopen(url)
    date = datetime.datetime.now().strftime("%d_%m_%Y_%H")
    out = open('vhi_id_%02d.csv' %index, 'wb')
    out.write(vhi_url.read())
    out.close()  
    os.rename('vhi_id_%02d.csv' %index, date+'_vhi_id_%02d.csv' %index)
    print ("VHI is downloaded...")
    
for index in range(1, 28):
    download(index)
```
